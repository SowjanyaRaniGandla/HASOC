```python
import re
df['text']=df.text.apply(lambda x:re.sub(r"[-()\"#/@;:<>{}`+=~|.!?,]", "", x))
df['text']=df.text.apply(lambda x:re.sub(r"[--]",' ',x))
df.head()
```

```python
import nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt')
stop=["के","का","एक","में","की","है","यह","और","से","हैं","को","पर","इस","होता","कि","जो","कर","मे","गया","करने","किया" ,"लिये","अपने","ने","बनी","नहीं ","तो","ही","या","एवं","दिया","हो","इसका","था","द्वारा","हुआ","तक","साथ","करना","वाले","बाद","लिए","आप","कुछ","सकते","किसी","ये","इसके","सबसे","इसमें","थे","दो","होने","वह","वे","करते","बहुत","कहा","वर्ग","कई","करें",
"होती","अपनी","उनके","थी","यदि","हुई","जा","ना","इसे","कहते","जब","होते","कोई","हुए","व","न","अभी","जैसे","सभी","करता","उनकी","तरह","उस","आदि","कुल","एस","रहा","इसकी","सकता","रहे","उनका","इसी","रखें","अपना","पे","उसके"]
def stopwrds(x):
    stop_words=set(stop)
    word_tokens=word_tokenize(x)
    sentences=[]
    for w in word_tokens:
        if w not in stop_words:
            sentences.append(w)
    return sentences
df['sp']=df.text.apply(stopwrds)
df.head()
```

```python
bad=[]
good=[]
with open("neg.txt","r",encoding="utf8") as f:
    for line in f.read().split('\n'):
        if line not in bad:
            bad.append(line)
with open("pos.txt","r",encoding="utf8") as f:
    for line in f.read().split('\n'):
        if line not in good:
            good.append(line)      
```
